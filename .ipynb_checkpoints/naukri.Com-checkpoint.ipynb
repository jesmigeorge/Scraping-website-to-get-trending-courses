## Manipulating the html codes to access the contents

```python
# Get element with tag name 'div'
element = driver.find_element(By.TAG_NAME, 'div')

    # Get all the elements available with tag name 'p'
elements = element.find_elements(By.CLASS_NAME, 'card-title')
for e in elements:
     print(e.text)
```

```python
courseList = driver.find_elements(By.CLASS_NAME, "courselist-card")
length = len(courseList)
courseTags  =[]
#s1= driver.find_elements(By.TAG_NAME, 'div')
# Get all the elements available with tag name 'p'
s2 = driver.find_elements(By.CLASS_NAME, 'tags-container')
#s3 = s2.find_element(By.TAG_NAME, "a")
#for e in elements:
# print(element.text)
for tags in s2:
    courseTags += tags.text.split('\n')
index=0
data=[]
for course in courseList:
    eachCourse = course.text.split("\n")
    courseBy,courseName,courseType,courseCertification = eachCourse[0],eachCourse[1],eachCourse[4],eachCourse[5]
    tags=courseTags[index]
    index+=1
    if courseType not in ['Easy','Beginner','Intermediate','Expert']:
        courseType = 'Not Specified'
    if courseCertification not in ['Certification','Diploma']:
        courseCertification = 'Not Specified'
    data.append((courseBy,courseName,courseType,courseCertification,tags))
 #   print(courseBy,courseName,courseReviews,courseLevel,courseCertification,tags)
   # print(course.text+"\n")
    #print(eachCourse)
#print(data)
```

### Using pandas to write the data scraped from website to a csv file

```python
df = pd.DataFrame(data, columns=['Course By', 'Course Name','Course Type','Course Certification','Course Tags']) 
df.to_csv('python_courses.csv', index=False, encoding='utf-8') 
```

# FINAL CODE

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
driver = webdriver.Chrome() 
url = "https://www.naukri.com/learning/python-courses-certification-training-st619-tg21"
driver.get(url)
courseList = driver.find_elements(By.CLASS_NAME, "courselist-card")
length = len(courseList)
courseTags  =[]
s2 = driver.find_elements(By.CLASS_NAME, 'tags-container')
for tags in s2:
    courseTags += tags.text.split('\n')
index=0
data=[]
for course in courseList:
    eachCourse = course.text.split("\n")
    courseBy,courseName,courseType,courseCertification = eachCourse[0],eachCourse[1],eachCourse[4],eachCourse[5]
    tags=courseTags[index]
    index+=1
    if courseType not in ['Easy','Beginner','Intermediate','Expert']:
        courseType = 'Not Specified'
    if courseCertification not in ['Certification','Diploma']:
        courseCertification = 'Not Specified'
    data.append((courseBy,courseName,courseType,courseCertification,tags))
df = pd.DataFrame(data, columns=['Course By', 'Course Name','Course Type','Course Certification','Course Tags'])  
df.to_csv('python_courses.csv', index=False, encoding='utf-8') 
driver.quit()
```
